# challenge 1

```python
import pandas as pd

books = pd.DataFrame({
    'title': ['The Pragmatic Programmer', 'Clean Code', 'Python Crash Course',
              'Data Science from Scratch', 'Designing Data-Intensive Apps'],
    'author': ['Hunt & Thomas', 'Robert Martin', 'Eric Matthes',
               'Joel Grus', 'Martin Kleppmann'],
    'genre': ['Tech', 'Tech', 'Tech', 'Data', 'Data'],
    'price': [45.99, 38.50, 29.99, 42.00, 55.00],
    'units_sold': [120, 95, 200, 80, 60],
    'in_stock': [True, True, False, True, False]
})

books
```

```python
books.info()
```

```python
books.describe()
```

```python
books['revenue'] = books['price'] * books['units_sold']
books
```

```python
# What is the total revenue across all books?
total_revenue = books['revenue'].sum()
total_revenue
```

```python
# 2. Use `.iloc` to retrieve the last 3 rows and first 3 columns.
books.iloc[-3:, :3]
```

```python
# 3. Filter books that are **in stock** AND have a **price above £40**.
# books.loc[books['in_stock'] & books['price']>40.0]
books.loc[books['in_stock'] & (books['price']>40.0)]
```

```python
# 4. Find the book with the **highest revenue** using `.loc` with `.idxmax()`.
highest_revenue_book = books.loc[books['revenue'].idxmax()]
highest_revenue_book
```

# challenge 3   

```python
# Use .apply() to add a price_tier column: 'Budget' (< £35), 'Standard' (£35–£50), 'Premium' (> £50).
def tiering(row):
    if row['price'] < 35:
        return 'Budget'
    elif 35 <= row['price'] <= 50:
        return 'Standard'
    else:
        return 'Premium'
    

books['price_tier']=books.apply(tiering, axis=1)
books
```

```python
# Sort the DataFrame by revenue descending.
books.sort_values('revenue', ascending=False, inplace=True)
books
```

```python
# Add a sales_rank column using .rank(ascending=False).
books['sales_rank'] = books['revenue'].rank(ascending=False)
books
```

```python
# Use .map() to encode genre as a number: {'Tech': 0, 'Data': 1}.
books['genre_code'] = books['genre'].map({'Tech':0,'Data':1})
books
```
